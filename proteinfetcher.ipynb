```python
import os
import pandas as pd
import re
```

```python
htdir = os.fsencode('/home/edvin/Documents/genome_analysis_2019/06_read_count/')
pdir = os.fsencode('/home/edvin/Documents/genome_analysis_2019/prokka/')


for file in os.listdir(htdir):
    #extract bin number
    bins = re.search('bin\d+',os.fsdecode(file)).group()
    
    #open file, read as pandas.
    count = os.fsdecode(htdir+file)
    df = pd.read_csv(count, sep='\t',names=[bins+'_gene','Oxic','Hypoxic'])
    #filter out lowly expressed stuff
    df_filtered = df[(df['Hypoxic']>5)|(df['Oxic']>5)]
    
    # loop through genes
    gffpath = os.fsdecode(pdir)+bins+'/PROKKA_05242019_noseq.gff'
    genes = df_filtered.iloc[:,0]
    
    for i in genes.index:
        gff = open(gffpath)
        for line in gff:  
            if re.search(genes.loc[i],line):
                protein = re.search("product=(.+$)",line).group(1)
                print(protein + ' found in ' + bins)
                df_filtered.loc[i,bins+'_gene'] = protein
                break
    df_filtered.to_csv(bins+'_geneexpression.csv')
    print("saved:"+bins)
        
```
